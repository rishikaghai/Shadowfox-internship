```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)
plt.title('Simple Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
plt.scatter(x, y, color='red')
plt.title("Scatter Plot - Matplotlib")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
import seaborn as sns
import pandas as pd

data = pd.DataFrame({'x': [1, 2, 3, 4, 5], 'y': [10, 20, 25, 30, 40]})
sns.lineplot(data=data, x='x', y='y')
plt.title("Line Plot - Seaborn")
plt.show()
```

```python
sns.scatterplot(data=data, x='x', y='y', color='red')
plt.title("Scatter Plot - Seaborn")
plt.show()
```

```python
categories = ['A', 'B', 'C']
values = [5, 7, 3]

plt.bar(categories, values)
plt.title("Bar Chart - Matplotlib")
plt.show()
```

```python
data = pd.DataFrame({'Category': ['A', 'B', 'C'], 'Value': [5, 7, 3]})
sns.barplot(data=data, x='Category', y='Value')
plt.title("Bar Chart - Seaborn")
plt.show()
```

```python
import numpy as np

data = np.random.randn(1000)
plt.hist(data, bins=30, edgecolor='black')
plt.title("Histogram - Matplotlib")
plt.show()
```

```python
sns.histplot(data, bins=30, kde=True)
plt.title("Histogram - Seaborn")
plt.show()
```

```python
sizes = [40, 30, 20, 10]
labels = ['A', 'B', 'C', 'D']

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Pie Chart - Matplotlib")
plt.show()
```

```python
import numpy as np

data = np.random.rand(5, 5)
sns.heatmap(data, annot=True, cmap='coolwarm')
plt.title("Heatmap - Seaborn")
plt.show()
```
